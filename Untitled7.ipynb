```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(5,100,6)
y1 = x**2
y2 = x**3
plt.plot(x,y1,x,y2)
plt.plot(x,y1,label="square")
plt.plot(x,y2,label="cube")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/Orders.csv")
print(df.head(4))
data = pd.DataFrame(df)
plt.figure(figsize=(20,8))
plt.bar(df['State'], df['City'], color='skyblue')
plt.xlabel('states')
plt.ylabel('cities')
plt.title('yearly sales')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Sample dataset: Student Scores
data = [60, 70, 80, 85, 90, 95, 100, 75, 85, 95, 80, 90, 100, 85, 95, 70, 65, 60, 90, 85]
# Create a figure with subplots
plt.figure(figsize=(15, 5))
# Bar Chart
plt.subplot(1, 3, 1)
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 12, 8, 14, 10]
sns.barplot(x=categories, y=values, palette="viridis")
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
# Histogram
plt.subplot(1, 3, 2)
plt.hist(data, bins=5, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Scores")
plt.ylabel("Frequency")
# Box Plot
plt.subplot(1, 3, 3)
sns.boxplot(data, color='red')
plt.title("Box Plot")
plt.xlabel("Scores")
```
